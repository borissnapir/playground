```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make some data with log link function and poisson distribution
a = 0.6
b = -0.4
X = np.random.uniform(low=1.0, high=5.0, size=100)
mu = np.exp(a*X+b) 
y = np.random.poisson(lam=mu)

fig, ax = plt.subplots()
ax.plot(X,y, marker='.', linestyle='')
```

With Poisson noise, the mean is equal to the variance, so the noise increases for higher values of y

```python
import statsmodels.api as sm
# add constant term to X to account for bias in linear regression
X_b = sm.add_constant(X)
mod = sm.GLM(y, X_b,
             family=sm.families.Poisson(link=sm.families.links.log()))
res = mod.fit()
```

```python
y_pred = res.predict(X_b)
idx = np.argsort(X)

fig, ax =plt.subplots()
ax.plot(X,y, marker='.', linestyle='')
ax.plot(X[idx], y_pred[idx], color="m", linewidth=3)
plt.show()
```

```python
# try with wrong probability distribution, ie gaussian instead of poisson
mod_g = sm.GLM(y, X_b,
             family=sm.families.Gaussian(link=sm.families.links.log()))
res_g = mod_g.fit()
```

```python
y_pred_g = res_g.predict(X_b)

fig, ax =plt.subplots()
ax.plot(X,y, marker='.', linestyle='')
ax.plot(X[idx], y_pred[idx], color="m", linewidth=3)
ax.plot(X[idx], y_pred_g[idx], color="c", linewidth=3)
plt.show()
```
