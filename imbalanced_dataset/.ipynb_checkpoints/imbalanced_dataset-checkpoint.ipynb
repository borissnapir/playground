```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator
from sklearn.metrics import f1_score

import tensorflow as tf
import tensorflow_addons as tfa
```

```python
n_classes = 4
X, y = make_classification(n_samples=10000, 
                              n_features=2,
                              n_redundant=0,
                              n_classes=n_classes,
                              weights=[0.96, 0.02, 0.01, 0.01],
                              class_sep=2,
                              n_clusters_per_class=1,
                              flip_y=0)
```

```python
fig, ax = plt.subplots()
for i in range(n_classes):
    select = y==i
    ax.plot(X[select,0], X[select,1], marker=".", linestyle='')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

# logistic regression with tensorflow

```python
class LogisticRegression(BaseEstimator):
    def __init__(self, n_classes, lr=0.001, loss='binary_crossentropy'):
        self.n_classes = n_classes
        self.lr = lr
        self.loss = loss
        
    # needed for grid search
    def get_params(self, deep=True):
        return {
            "lr": self.lr,
            'loss': self.loss
        }
        
    # needed for grid search
    def set_params(self, **parameters):
        for parameter, value in parameters.items():
            setattr(self, parameter, value)
        # return self to allow chaining
        return self
    
    def build_model(self, lr):
        self.model = tf.keras.models.Sequential([tf.keras.layers.Dense(self.n_classes, activation='softmax')])
        self.optimizer = tf.keras.optimizers.SGD(learning_rate=lr, momentum=0.9)       
        self.model.compile(loss=self.loss, optimizer=self.optimizer, metrics=['accuracy'])

    # fit model
    def fit(self, X_train, Y_train, XY_valid=None, epochs=50, **kwargs):
        self.build_model(lr=self.lr)
        if XY_valid is not None:
            callbacks = [tf.keras.callbacks.EarlyStopping(
                monitor='val_loss', 
                patience=5,
            )]
        else:
            callbacks = []
        self.history = self.model.fit(X_train, Y_train,
                                      epochs=epochs,
                                      validation_data=XY_valid,
                                      verbose=1,
                                      callbacks=callbacks)
        return self
                  
    def predict_proba(self, X):
        return self.model.predict(X)
    
    def predict(self, X):
        return (self.predict_proba(X)>=0.5).astype(int)
    
    def score(self, X, y):
        y_pred = self.predict(X)
        return f1_score(y_pred=y_pred, y_true=y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=0)
```

**crossentropy**

```python
model = LogisticRegression(loss="sparse_categorical_crossentropy", n_classes=n_classes)
```

```python
model.fit(X_train, y_train, XY_valid=(X_val, y_val), epochs=500)
```

```python
y_pred = model.predict(X_test)
```

```python
confusion_matrix(y_test, np.argmax(y_pred, axis=-1))
```

**balanced focal crossentropy loss**

```python
focal_loss = tfa.losses.SigmoidFocalCrossEntropy(
    from_logits=False,
    alpha=0.01,
    gamma=2.0,
    name='sigmoid_focal_crossentropy'
)
```

```python
model_focal = LogisticRegression(
    loss=focal_loss
)
```

```python
model_focal.fit(X_train, y_train, XY_valid=(X_val, y_val), epochs=500)
```

```python
y_pred = model_focal.predict(X_test)
confusion_matrix(y_test, y_pred)
```
