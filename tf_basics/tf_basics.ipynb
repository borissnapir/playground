```python
import tensorflow as tf
import numpy as np
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt
```

```python
# Load sample images
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255

fig, ax = plt.subplots(1 ,2 , figsize=(12,4))
ax[0].imshow(china)
ax[1].imshow(flower)
plt.show()
```

```python
images = np.array([china, flower])
batch_size, height, width, channels = images.shape
```

# Simple 2d conv in tensorflow with manually defined kernel

https://www.tensorflow.org/api_docs/python/tf/nn/conv2d

```python
# Create 2 filters
filters = np.zeros(shape=(7, 7, channels, 2), dtype=np.float32)
filters[:, 3, :, 0] = 1 # vertical line
filters[3, :, :, 1] = 1 # horizontal line
```

```python
outputs = tf.nn.conv2d(images, filters, strides=1, padding="SAME")
print(f"Output size: {outputs.shape}")

fig, ax = plt.subplots(1 ,2 , figsize=(12,4))
ax[0].imshow(outputs[0, :, :, 1], cmap="gray") # plot 1st image's 2nd feature map
ax[1].imshow(outputs[1, :, :, 1], cmap="gray") # plot 1st image's 2nd feature map
plt.show()
```
